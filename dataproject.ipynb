```python
from imblearn.over_sampling import SMOTE

cols=['score','critics','users','multi',
'cooperative','publisher','global_sales',
'platform_3DS','platform_DS','platform_PC',
'platform_PSP','platform_Wii','platform_WiiU',
'genre_Action','genre_Adventure','genre_Fighting',
'genre_Misc','genre_Platform','genre_Puzzle',
'genre_Racing','genre_Role-Playing',
'genre_Shooter','genre_Simulation',
'genre_Sports','genre_Strategy']
X=df8[cols]
y=df8['cluster']

X_sample,y_sample=SMOTE(random_state=0).fit_resample(X, y)

X_samp = pd.DataFrame(data=X_sample,columns=cols )
y_samp = pd.DataFrame(data=y_sample,columns=['cluster']) 

sampdf=pd.concat([X_samp,y_samp],axis=1)


```

```python
sampdf["cluster"].value_counts()
```

```python
X=X_samp[cols]
y=y_samp['cluster']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, 
test_size=0.2, stratify=y, random_state=0)
```

```python
mean_on_train = X_train.mean(axis=0)
std_on_train = X_train.std(axis=0)

X_train_scaled = (X_train - mean_on_train) / std_on_train 
X_test_scaled = (X_test - mean_on_train) / std_on_train 
```

```python
from sklearn.linear_model import LogisticRegression 
model = LogisticRegression(C=1000, random_state=0) 
model
model.fit(X_train_scaled, y_train) 
```

```python
import numpy as np

pred=model.predict(X_test_scaled)
print("학습용 :",model.score(X_train_scaled, y_train)) 
print("검증용 :",model.score(X_test_scaled, y_test)) 
print("검증용 :",np.mean(pred == y_test))
```

```python
import seaborn as sns 
from sklearn.metrics import confusion_matrix

pred=model.predict(X_test_scaled)
cm=confusion_matrix(y_test, pred) 
df_cm=pd.DataFrame(cm, index=[0, 1], columns=[0, 1]) 
sns.set(font_scale=1.4) 
plt.figure(figsize=(9,6))
plt.title("Confusion Matrix")
sns.heatmap(df_cm, annot=True) 
```

```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier

train_rate = [] 
test_rate = [] 
neighbors = range(1, 11)

for n in neighbors: 
    knn = KNeighborsClassifier(n_neighbors=n) 
    knn.fit(X_train_scaled, y_train) 
    train_rate.append(knn.score(X_train_scaled, y_train)) 
    test_rate.append(knn.score(X_test_scaled, y_test))

max_rate=max(test_rate) 
idx=np.argmax(test_rate)
print('최적의 k:',idx+1)
```
